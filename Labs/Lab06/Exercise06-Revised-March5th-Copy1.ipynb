You are given 5 samples of 100 rolls of a die. The die is chosen at random from one of two dice, either of which may be biased. Given the following counts, estimate the parameters for each die (i.e. the probabilites of getting a 1, 2, 3, 4, 5 or 6).

````python
df = pd.read_csv('dice.csv', index_col=0)
df
````

````python
### Answer Q2###

#np.array([df])
# Below are 5 samples, each with 100 rolls
xs = np.array([(2, 14, 10, 20, 26, 28), (1, 13, 12, 17, 21, 36), (21, 18, 16, 12, 14, 19), (15, 20, 20, 15, 12, 18), (5, 10, 11, 24, 17, 33)])

# Starting points for the biased and unbiased coin respectively
thetas = np.array([[0.3, 0.12, 0.11, 0.20, 0.21, 0.33], [1.0/6,1.0/6,1.0/6,1.0/6,1.0/6,1.0/6]])

tol = 0.01
max_iter = 100

ll_old = 0

# The main part of this code is citated from the lecture notes
for i in range(max_iter):
    ws_A = []
    ws_B = []
    
    vs_A = []
    vs_B = []
    
    ll_new = 0

    # E-step: calculate probability distributions over possible completions
    for x in xs:

        # multinomial (binomial) log likelihood
        ll_A = np.sum([x*np.log(thetas[0])])
        ll_B = np.sum([x*np.log(thetas[1])])
            
        # [EQN 1]
        denom = np.exp(ll_A) + np.exp(ll_B)       
        w_A = np.exp(ll_A)/denom
        w_B = np.exp(ll_B)/denom
        
        ws_A.append(w_A)
        ws_B.append(w_B)

        # used for calculating theta
        vs_A.append(np.dot(w_A, x))
        vs_B.append(np.dot(w_B, x))
        
        # update complete log likelihood
        ll_new += w_A * ll_A + w_B * ll_B

    # M-step: update values for parameters given current distribution
    # [EQN 2]
    thetas[0] = np.sum(vs_A, 0)/np.sum(vs_A)
    thetas[1] = np.sum(vs_B, 0)/np.sum(vs_B)
    # print distribution of z for each x and current parameter estimate
    
    if np.abs(ll_new - ll_old) < tol:
        break
    ll_old = ll_new       

print 'The parameter for the biased coin is: '
print thetas[0]
print 'The parameter for the unbiased coin is: '
print thetas[1]
````

**Exercise 3 (10 points)**

You have some data about the intervals between arrivals (in minutes) of customers to a bank. You want to know how many customers to expect from 9 AM to 5 PM. Simulate 1000 sequences of customer arrivals over an 8 hour period and esitmate the mean and variance. 

````python
### Answer Q3###

from statsmodels.distributions.empirical_distribution import ECDF

data = np.loadtxt('arrival_intervals.txt')

from scipy.interpolate import interp1d

# The main part of this code is citated from the lecture notes
def extrap1d(interpolator):
    """From StackOverflow http://bit.ly/1BjyRfk"""
    xs = interpolator.x
    ys = interpolator.y

    def pointwise(x):
        if x < xs[0]:
            return ys[0]+(x-xs[0])*(ys[1]-ys[0])/(xs[1]-xs[0])
        elif x > xs[-1]:
            return ys[-1]+(x-xs[-1])*(ys[-1]-ys[-2])/(xs[-1]-xs[-2])
        else:
            return interpolator(x)

    def ufunclike(xs):
        return np.array(map(pointwise, np.array(xs)))

    return ufunclike


x = np.concatenate([data])

# Build the empirical distribution
ecdf = ECDF(x)
inv_cdf = extrap1d(interp1d(ecdf.y, ecdf.x, 
                            bounds_error=False, assume_sorted=True))

# This is to store the # of customers in each of the simulations 
n = 1000

# make a list of n 0s
p = np.zeros(n)

# Do the simulation for 1000 times
for i in range(n):
    r = np.random.uniform(0, 1, 100)
    ys = inv_cdf(r)
    # Count the # of customers that can arrive betwen 9 and 5 for this simulation
    count = ys[np.cumsum(ys)<=480]
    # store the count in the i_th element of p 
    p[i] = len(count)

print 'The mean and variance for the number of customer is:'
print(np.mean(p),np.var(p))
````

The Cauchy distribution is given by

````python
### Answer Q4 Part 1###

# A
n = 1000
y = stats.cauchy().rvs(n)
h_mc = 1.0/n * np.sum(y > 2)
print 'Tail probability P(X>2) Sampling from the Cauchy distribution directly:'
print h_mc
print '\n'

#B
n = 1000
y = stats.uniform().rvs(n)
h_cv = 1.0/n * np.sum(2.0/(np.pi * (4 + y**2)))
print 'Sampling from the uniform distribution using an appropriate change of variables:'
print h_cv
print '\n'
print 'Need more samples to reach the higher precision. 1,000 samples can just guarantee the precision for the firsrt digit.'
````

````python
### Answer Q4 Part 2###

#A sampling from the Cauchy distribution using mulitple Monte Carlo sequences
from scipy.stats import cauchy

n = 1000
reps = 1000


x = stats.cauchy().rvs((n, reps))
# Define which random samples are > 2 for 1,000 MC sequence
y = 1/np.arange(1, n+1)[:, None] * np.cumsum(x  > 2, axis=0)
upper, lower = np.percentile(y, [2.5, 97.5], axis=1)

plt.axis([0,1000,-0.2,0.4])
plt.plot(np.arange(1, n+1), y, c='grey', alpha=0.02)
plt.plot(np.arange(1, n+1), y[:, 0], c='red', linewidth=1);
plt.plot(np.arange(1, n+1), upper, 'b', np.arange(1, n+1), lower, 'b-');
plt.show()

print 'We can see that the P(X>2) traverse the space well and convergence to the theoretical value of 0.1486.'
````

**Transform of Variables**
$$
f(x) = \frac{1}{\pi (1 + x^2)}, \ \ -\infty \lt x \lt \infty 
$$

We need estimate $P(X>2)$
$$
f(x) = \frac{1}{\pi (1 + x^2)}, x \geq 2
$$

Change of Variable:
$$
y = \frac{2}{x} 
$$

Therefore, $0 < y \leq 1 $

$$
x=\frac{2}{y}, \frac{dx}{dy}=-2y^{-2}
$$

Finally, we can get:
$$
f(y) = \frac{2}{\pi (4 + y^2)},0 < y \leq 1 
$$

````python
# B sampling from the uniform distribution using bootstrap samples 
n = 1000
reps = 1000

# Define the function that we just did the tranform of variable
def f(x):
    return 2.0/(np.pi*(4.0 + x**2))

# Do the bootstrap sampling
x = f(np.random.random((n, reps)))
xb = np.random.choice(x[:,0], (n, reps), replace=True)
yb = 1/np.arange(1, n+1)[:, None] * np.cumsum(xb, axis=0)
upperb, lowerb = np.percentile(yb, [2.5, 97.5], axis=1)

plt.plot(np.arange(1, n+1)[:, None], yb, c='grey', alpha=0.02)
plt.plot(np.arange(1, n+1), yb[:, 0], c='red', linewidth=1)
plt.plot(np.arange(1, n+1), upperb, 'b', np.arange(1, n+1), lowerb, 'b')
plt.show()

print 'We can see that the P(X>2) traverse the space well and convergence to the theoretical value of 0.1486.'
````
