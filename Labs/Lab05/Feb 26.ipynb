Background for Exercises 1 and 2
----

The first 2 exercises are about using Newton's method to find the cube roots of unity - find $z$ such that $z^3 = 1$. From the fundamental theorem of algebra, we know there must be exactly 3 complex roots since this is a degree 3 polynomial.

We start with Euler's fabulosu equation
$$
e^{ix} = \cos x + i \sin x
$$

Raising $e^{ix}$ to the $n$th power where $n$ is an integer, we get from Euler's formula with $nx$ substituting for $x$
$$
(e^{ix})^n = e^{i(nx)} = \cos nx + i \sin nx
$$

Whenever $nx$ is an integer multiple of $2\pi$, we have
$$
\cos nx + i \sin nx = 1
$$

So
$$
e^{2\pi i \frac{k}{n}}
$$
is a root of 1 whenever $k/n = 0, 1, 2, \ldots$.

So the cube roots of unity are $1, e^{2\pi i/3}, e^{4\pi i/3}$. 

While we can do this analytically, the idea is to use Newton's method to find these roots, and in the process, disover some rather perplexing behavior of Newton's method.

**Exercise 1 (10 points)**. Netwon's method for functions of complex variables - stability and basins of attraction.

1. Write a function with the following function signature `newton(z, f, fprime, max_iter=100, tol=1e-6)` where
    - `z` is a starting value (a complex number e.g.  ` 3 + 4j`)
    - `f` is a function of `z`
    - `fprime` is the derivative of  `f`
The function will run until either max_iter is reached or the absolute value of the Newton step is less than tol. In either case, the function should return the number of iterations taken and the final value of `z` as a tuple (`i`, `z`). 

2. Define the function `f` and `fprime` that will result in Newton's method finding the cube roots of 1. Find 3 starting points that will give different roots, and print both the start and end points. 

````python
# Q1: Write a function with the following function signature newton(z, f, fprime, max_iter=100, tol=1e-6)
# Pre: The function will take these parameters:
#                            1. the starting point z, a complex number;
#                            2. f is a function of x
#                            3. fprime is the derivative of f
# Post: Return the number of iterations taken and the final value of z as a tuple (i, z).
def newton(z, f, fprime, max_iter=100, tol=1e-6):
    count = 0
    diff = 1 #setup a initial value (large enough)
    znew = complex(z)
    # The required max iteration time:
    while count < max_iter:
        diff = - f(znew)/fprime(znew)
        znew = expand(znew + diff, complex=True)
        # The while loop will break if the error term is small enough
        if np.absolute(diff) <= tol:
            break
        # count the number of iteration the algorithm take before breaking
        count += 1
    # return the final tuple (i, z)
    result = (count, znew)
    return result

# Q2 Define the function f and fprime that will result in Newton's method 
def f(z):
    return np.power(z,3) - 1.0

# fprime is the first order derivative of f
def fprime(z):
    return 3*np.power(z,2)

print 'Starting point: -0.3, and the number of iterations for being convergence, the convergence points are listed below:'
print newton(0.3, f, fprime), '\n'

print 'Starting point 1: -0.3 + i, and the number of iterations for being convergence, the convergence points are listed below:'
print newton(-0.3 + I, f, fprime), '\n'

print 'Starting point 2: -0.3 - i, and the number of iterations for being convergence, the convergence points are listed below:'
print newton(-0.3 - I, f, fprime), '\n'


print "The three points (0.3, -0.3 + I, -0.3 - I)above gives 3 different convergence points: 1, -0.5 + 0.866025403784439*I, "
print "-0.5 - 0.866025403784439*I." 
print "The algorithm converges really fast: the most inefficient one takes 8 steps to converge."
print "The three convergence points can be discovered through the plot below in the next question. " , '\n'
````

**Exercise 2 (10 points)**. Write the following two plotting functions to see some (pretty) aspects of Newton's algorithm in the complex plane.

1. The first function `plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv')` calculates and stores the number of iterations taken for convergence (or max_iter) for each point in a 2D array. Make a plot for the cube roots of 1 - since there are 3 roots, there should be only 3 colors in the plot.

````python
from matplotlib.pyplot import *
from colorsys import hls_to_rgb
# Plot the array "A" using colorize
import pylab as plt

from PIL import Image
import cmath


print 'In order to make the algorithm run faster, I need do a little adjustment on the code I wrote for the first'
print "problem. I want the complex number that is returned by newton be not expanded. Meanwhile, the whole algorithm is"
print "the same for the newton I wrote above."

# the f, fprime and newton are the revised algorithm above
def f(z):
    return z**3 - 1.0

def fprime(z):
    step = 1e-6
    return (f(z + complex(step, step)) - f(z)) / complex(step, step)

def newton1(z, f, fprime, max_iter=100, tol=1e-6):
    count = 0
    diff = 1 #setup a initial value (large enough)
    znew = z
    for count1 in range(max_iter):
        diff = f(z)/fprime(z)
        znew = znew - diff
        if(np.absolute(diff) <= tol):
            # the result will return the raw imaginary number directly
            zz = round(znew.real,2)+ round(znew.imag,2)
            zz = znew
            count = count + 1
            break  
        count = count + 1
    result = (count, znew)
    return result

# Question 1 calculates and stores the number of iterations taken for convergence (or max_iter) for each point in a 2D array

# This function is to plot the Newton fractal at an extent=[-1,1,-1,1]
def plot_newton_iters(f, fprime, n=1, extent=[-1,1,-1,1], cmap='hsv'):
    # Define a matrix A, the x direction is the real axis, and the y direction is the imaginary axis
    A = 1.0*np.zeros((n,n),dtype='complex')
    # Divide the real and imaginary axis into n parts evenly
    axis_x = np.linspace(-1.0,1.0,n)*float(1.0)
    axis_y = np.linspace(-1.0,1.0,n)*float(1.0)
    # mash grid the two axis
    X,Y = np.meshgrid(axis_x,axis_y)
    # define Z to be a complex number that is built by X and Y
    Z = X + Y*1j
    for i in np.arange(n):
        for j in np.arange(n):
            # This is to pass the complex number within each grid through the fnction newton1, the function I 
            # just revised
            A[i,j] = newton1(complex(float(Z[i,j].real),float(Z[i,j].imag)), f, fprime, max_iter=100, tol=1e-6)[1]
            # return the plot
    return plt.imshow(colorize(A), interpolation='none', cmap = 'hsv' , extent = [-1,1,-1,1])

plot_newton_iters(f, fprime, n=200, extent=[-1,1,-1,1], cmap='hsv')
````

1. Stochastic gradient descent.  In this method, gradient descent is used essentially by fitting *one data point at a time*.  Recall the usual gradient descent step:

$$\beta_{i+1} = \beta_i - \nabla \ell(\beta_i)$$

where

$$\nabla \ell(\beta_i) = \sum\limits_{j=1}^n \nabla\ell(\beta_i,x_j,y_j)$$

and $\ell$ is the log-likelihood function.  All of the data is used to make the next step toward the optimal $\beta$.  In stochastic gradient descent, only one point at a time is used to determine the next $\beta$:

$$\beta_{i+1} = \beta_i - \alpha\nabla \ell(\beta_i,x_j,y_j)$$

where $\alpha$ is the step size.  For simplicity, we'll take a constant $\alpha=1$.

   a. Shuffle data points (i.e. randomly permute the order of the $(x_j,y_j)$
